Let's combine the llm-generated data into one big llm-generated/phishing.csv file

```python
import pandas as pd

# Load the CSV files
df1 = pd.read_csv("../dataset/llm-generated/copilot_generated1.csv")
df2 = pd.read_csv("../dataset/llm-generated/copilot_generated2.csv")

# Combine them
combined_df = pd.concat([df1, df2], ignore_index=True)

# Save the result
combined_df.to_csv("../dataset/llm-generated/copilot_generated.csv", index=False)
```

```python
import pandas as pd

# Generated datasets
chatgpt = pd.read_csv("../dataset/llm-generated/chatgpt_generated.csv")
claude = pd.read_csv("../dataset/llm-generated/claude_generated.csv")
copilot = pd.read_csv("../dataset/llm-generated/copilot_generated.csv")
grok = pd.read_csv("../dataset/llm-generated/grok_generated.csv")

for df in [chatgpt, claude, copilot, grok]:
    df["clean_text"] = (
        df["clean_text"]
        .astype(str)
        .str.replace(r"\r\n|\n|\r", " ", regex=True)
        .str.replace(r"\s+", " ", regex=True)
        .str.strip()
    )

# --------------------------
# Combine
# --------------------------

combined = pd.concat(
    [chatgpt, claude, copilot, grok],
    ignore_index=True
)

# --------------------------
# Save
# --------------------------

combined.to_csv(
    "../dataset/llm-generated/phishing.csv",
    index=False
)

print(f"Final rows: {len(combined):,}")
print(combined["label"].value_counts())
print(combined["source"].value_counts())
```
